```python
import os,numpy as np,pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
path=os.getcwd()+"/../Models/EmbryonClassif/embryon_classif/"
print(path)
f=os.listdir(path)

a=f[2]
print(a)
```
