```python
import os
import pandas as pd
import imageio as im
import numpy as np
import matplotlib.pyplot as plt
```

```python
path=os.getcwd()+"/../"
pathtoFrame=path+"EmbryonVideoClassif/Frames/"
pathtocsv=path+"/DataSplit/"
pathtotriplet=path+"Triplet/Frames/"
```

```python
os.listdir(pathtocsv)
```

```python
df=pd.DataFrame()
for i in os.listdir(pathtocsv):
    if "Binary" not in i and "Raw" not in i:
        print (i)
        df1=pd.read_csv(pathtocsv+i)
        df=pd.concat([df,df1])
```

```python
classes=pd.unique(df['Class'])
for t, row in df.iterrows():
    for i in classes:
        if i == row['Class']:
            Img=im.imread(pathtoFrame+"/../../"+row['Image'])
            #print(os.listdir(pathtotriplet+str(i)))
            cpath=pathtotriplet+str(i)+"/"
            print(os.path.isdir(cpath+(str(row['Image']).split("/")[2])))
            if os.path.isdir(cpath+(str(row['Image']).split("/")[2])) :
                print(cpath+str(row['Image']).split("/")[2]+"__exist")
            else:
                os.mkdir(cpath+(str(row['Image']).split("/")[2]))
            print(cpath+(str(row['Image']).split("/")[2])+"/"+(str(row['Image']).split("/")[3]))
            im.imwrite(cpath+(str(row['Image']).split("/")[2])+"/"+(str(row['Image']).split("/")[3]),Img)
            break
    
            

            
```
